```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="/content/drive/MyDrive/trash-classification-YOLO/dataset/data.yaml",
    epochs=50,
    lr0=0.01,
    freeze=22,
    patience=20,
    imgsz=640,
    batch=16,
    save=True,
    project="/content/drive/MyDrive/trash-classification-YOLO/yoloV8nFineTune1",
    name="train1",
    exist_ok=True,

    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    flipud=0.5,
    fliplr=0.5,
    degrees=10,
    translate=0.10,
    scale=0.20,
    shear=0.10,
    perspective=0.0005,
    mixup=0.15,
    mosaic=1.0,
)
```
